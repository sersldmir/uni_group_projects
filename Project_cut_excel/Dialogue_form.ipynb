```python
print('''Приветствуем Вас в нашей программе по работе с файлами разных типов!
        Для начала работы необходимо загрузить готовый файл или сразу несколько (тогда они будут преобразованны в единую структуру данных)
        Для этого введите "Загрузка"
        Также вы можете создать базу данных прямо в программе. Для этого введите "Создать".
        ''')
while True:
    a=input('Введите ответ:\n')
    if a=='Загрузка' or a=='Создать':
        break
    else:
        print('Не то ввели!')
if a=='Создать':
    d=create_tabledata()
    while True:
        a=input('Хотите автоматически определить тип данных и подогнать значения под них? Да или Нет\n')
        if a=='Да':
            dic=get_column_types(d,by_number=False)
            load_w_ty(d,dic)
            dd=1
            break
        elif a=='Нет':
            break
        else:
            print('Неверный ответ!')
elif a=='Загрузка':
    while True:
        a=input('Выбирайте формат (поддерживаемые форматы: csv, pickle и txt)\n')
        if a=='csv' or a=='pickle' or a=='txt':
            break
        else:
            print('Неверный формат!')
    if a=='csv':
        files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
        d=load_tablecsv(*files.split())
    elif a=='pickle':
        files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
        d=load_tablepickle(*files.split())
    elif a=='txt':
        files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
        d=load_tabletxt(*files.split())
    while True:
        a=input('Хотите автоматически определить тип данных и подогнать значения под них? Да или Нет\n')
        if a=='Да':
            dic=get_column_types(d,by_number=False)
            load_w_ty(d,dic)
            dd=1
            break
        elif a=='Нет':
            break
        else:
            print('Неверный ответ!')
print('''Теперь программа имеет доступ к данным! Вот какие манипуляции вы можете провести:
- вывести строки таблицы по их индексу
- вывести строки таблицы по первому значению столбцов
- получить словарь типа "Имя столбца/номер столбца: Тип значений" и изменить его
- получить/установить значения столбцов
- вывести таблицу на печать
- сохранить таблицу как одну или как несколько
- склеить две таблицы
- разделить таблицу
- проводить арифметические операции над столбцами (сложение, вычитание, деление и умножение)
- с помощью логических операций (равно/не равно, больше/меньше, больше или равно/меньше или равно) получить таблицу с рядами, для которых логическая операция верна
''')
while True:
    try:
        a=input('''Так что хотите?:
        - вывести строки таблицы
        - получить словарь типов с последующим изменением
        - получить/установить значения столбцов
        - вывести таблицу на печать
        - сохранить таблицу как одну или как несколько
        - склеить две таблицы
        - разделить таблицу
        - проводить арифметические операции над столбцами
        - таблица из логики
        - завершить программу
        Скопируйте одну из опций в поле ввода''')
        if a=='вывести строки таблицы':
            while True:
                an=input('По номеру строки или значению первого столбца?\n')
                if an=='По номеру строки' or an=='По значению первого столбца':
                    break
                else:
                    print('Не понял')
            if an=='По номеру строки':
                while True:
                    try:
                        start=int(input('Введите начало интервала вывода\n'))
                    except ValueError:
                        print('Ввели не (целое) число!')
                    else:
                        break
                while True:
                    try:
                        stop=int(input('Введите конец интервала вывода. Если хотите вывести только одну строчку, то введите 0\n'))
                    except ValueError:
                        print('Ввели не (целое) число!')
                    else:
                        break
                while True:
                    con=input('Вы хотите сохранить результаты в отдельную таблицу (введите "Сохранить") или просто вывести их на экран (введите "Вывести")?\n')
                    if con=='Cохранить':
                        con=True
                        break
                    elif con=='Вывести':
                        con=False
                        break
                    else:
                        print('Не то ввели!')
                get_rows_by_number(d,start,stop,con)
                continue
            elif an=='По значению первого столбца':
                lst=input('Введите значения первого столбца через пробел\n').split()
                while True:
                    con=input('Вы хотите сохранить результаты в отдельную таблицу (введите "Сохранить") или просто вывести их на экран (введите "Вывести")?\n')
                    if con=='Cохранить':
                        con=True
                        break
                    elif con=='Вывести':
                        con=False
                        break
                    else:
                        print('Не то ввели!')
                get_rows_by_index(d,*lst,con)
                continue
        elif a=='получить словарь типов с последующим изменением':
            if dd==1:
                print(dic)
            else:
                while True:
                    an=input('Если вы хотите словарь вида "имя: тип", то введите "имя", а если "номер столбца: тип", то введите "номер"\n')
                    if an=='имя' or an=='номер':
                        break
                    else:
                        print('Не то ввели')
                    if an=='имя':
                        con=False
                    elif an=='номер':
                        con=True
                dic=get_column_types(d,con)
                print(dic)
            while True:
                an=input('Хотите его поменять? Да или Нет\n')
                if an=='Да' or an=='Нет':
                    break
                else:
                    print('Не то ввели')
            if an=='Да':
                dic=set_column_types(d,dic,con)
                print('Измененный словарь')
                print(dic)
                while True:
                    an=input('Хотите применить изменения к таблице? Да или Нет\n')
                    if an=='Да' or an=='Нет':
                        break
                    else:
                        print('Не то ввели')
                if an=='Да':
                    load_w_ty(d,dic)
                    continue
                if an=='Нет':
                    continue
        elif a=='получить/установить значения столбцов':
            while True:
                try:
                    column=int(input('Введите колонку, с которой хотите работать\n'))
                except ValueError:
                    print('Ввели не (целое) число!')
                else:
                    break
            if len(d)==2:
                get_value(d,column)
                while True:
                    an=input('Хотите поменять значения в этом столбце? Да или Нет\n')
                    if an=='Да' or an=='Нет':
                        break
                    else:
                        print('Не то ввели')
                if an=='Да':
                    lst=input('Введите желаемое значение')
                    set_value(d,lst,column)
                    continue
                if an=='Нет':
                    continue
            else:
                get_values(d,column)
                while True:
                    an=input('Хотите поменять значения в этом столбце? Да или Нет\n')
                    if an=='Да' or an=='Нет':
                        break
                    else:
                        print('Не то ввели')
                if an=='Да':
                    lst=input('Введите желаемые значения через пробел\n').split()
                    set_values(d,lst,column)
                    continue
                if an=='Нет':
                    continue
        elif a=='вывести таблицу на печать':
            print_table(d)
            continue
        elif a=='склеить две таблицы':
            print('Загрузите вторую таблицу!')
            while True:
                a=input('Выбирайте формат (поддерживаемые форматы: csv, pickle и txt)\n')
                if a=='csv' or a=='pickle' or a=='txt':
                    break
                else:
                    print('Неверный формат!')
            if a=='csv':
                files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
                d1=load_tablecsv(*files.split())
            elif a=='pickle':
                files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
                d1=load_tablepickle(*files.split())
            elif a=='txt':
                files=input('Введите файл(ы), которые хотите загрузить через пробел\n')
                d1=load_tabletxt(*files.split())
            while True:
                a=input('Хотите автоматически определить тип данных и подогнать значения под них? Да или Нет\n')
                if a=='Да':
                    dic1=get_column_types(d1,by_number=False)
                    load_w_ty(d1,dic1)
                    break
                elif a=='Нет':
                    break
                else:
                    print('Неверный ответ!')
            print('Вторая таблица загружена! Начинаем слияние!')
            concat(d,d1)
            continue
        elif a=='разделить таблицу':
            while True:
                try:
                    row_number=int(input('Введите номер строки-разделителя\n'))
                except ValueError:
                    print('Ввели не (целое) число!')
                else:
                    break
            split(d,row_number)
            continue
        elif a=='проводить арифметические операции над столбцами':
            while True:
                op=input('Скопируйте название операции: плюс, минус, деление, умножение\n')
                if op=='плюс' or op=='минус' or op=='деление' or op=='умножение':
                    break
                else:
                    print('Не то ввели')
            if op=='плюс':
                add(d)
                continue
            elif op=='минус':
                sub(d)
                continue
            elif op=='деление':
                div(d)
                continue
            elif op=='умножение':
                mul(d)
                continue
        elif a=='таблица из логики':
            while True:
                op=input('Скопируйте название операции: больше, меньше, меньше или равно, больше или равно, равно, не равно\n')
                if op=='больше' or op=='меньше' or op=='меньше или равно' or op=='больше или равно' or op=='равно' or op=='не равно':
                    break
                else:
                    print('Не то ввели')
            if op=='больше':
                lst=gr(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif an=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
            elif op=='меньше':
                lst=ls(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif a=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
            elif op=='меньше или равно':
                lst=le(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif an=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
            elif op=='больше или равно':
                lst=ge(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif an=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
            elif op=='равно':
                lst=eq(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif an=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
            elif op=='не равно':
                lst=ne(d)
                print('Вот что получилось')
                print(lst)
                while True:
                    an=input('Хотите преобразовать лист в таблицу? Да или Нет\n')
                    if an=='Да':
                        while True:
                            an=input('Хотите сохранить результат в отдельную таблицу? Да или Нет\n')
                            if an=='Да':
                                copy_table=True
                                break
                            elif an=='Нет':
                                copy_table=False
                                break
                            else:
                                print('Неверный ответ!')
                        filter_rows(d,lst,copy_table)
                        break
                    elif an=='Нет':
                        break
                    else:
                        print('Неверный ответ!')
                continue
        elif a=='сохранить таблицу как одну или как несколько':
            while True:
                an=input('Выбирайте формат (поддерживаемые форматы: csv, pickle и txt)\n')
                if an=='csv' or an=='pickle' or an=='txt':
                    break
                else:
                    print('Неверный формат!')
            if an=='csv':
                while True:
                    try:
                        max_rows=int(input('По сколько строк вы хотите разбить данные? Если хотите сохранить целиком, то введите 0\n'))
                    except ValueError:
                        print('Ввели не (целое) число!')
                    else:
                        break
                save_tablecsv(d,max_rows)
                continue
            elif an=='pickle':
                while True:
                    try:
                        max_rows=int(input('По сколько строк вы хотите разбить данные? Если хотите сохранить целиком, то введите 0\n'))
                    except ValueError:
                        print('Ввели не (целое) число!')
                    else:
                        break
                save_tablepickle(d,max_rows)
                continue
            elif an=='txt':
                while True:
                    try:
                        max_rows=int(input('По сколько строк вы хотите разбить данные? Если хотите сохранить целиком, то введите 0\n'))
                    except ValueError:
                        print('Ввели не (целое) число!')
                    else:
                        break
                save_tabletxt(d,max_rows)
                continue
        elif a=='завершить программу':
            raise KeyboardInterrupt
        else:
            print('Не то ввели')
    except KeyboardInterrupt:
        print('Спасибо за работу со мной! Приходите ещё!')
        break
        
    
```
