```python
#Общий класс фигур + поле
class Figure(object):
    board = [['.']*8 for _ in range(8)]
    board.insert(0, ' ABCDEFGH')
    board.append(' ABCDEFGH')
    for i in range(1, 9):
        board[i].insert(0, 9-i)
        board[i].append(9-i)
    white_kill=[]
    black_kill=[]
    massive=[]
    
    @classmethod
    def print_board(cls):
        for i in cls.board:
            for k in i:
                print(k, end=' ')
            print()
        print(f'Потери белых - {cls.white_kill}',f'Потери черных - {cls.black_kill}', sep='\n')
        
            
    @classmethod
    def field_init(cls):
        for i in range(1,len(cls.board)):
            for g in range(1,len(cls.board[i])-1):
                if i==1:
                    if g==1 or g==8:
                        cls.board[i][g]='r'
                    elif g==2 or g==7:
                        cls.board[i][g]='n'
                    elif g==3 or g==6:
                        cls.board[i][g]='b'
                    elif g==4:
                        cls.board[i][g]='q'
                    elif g==5:
                        cls.board[i][g]='k'
                elif i==2:
                    cls.board[i][g]='p'
                elif i==7:
                    cls.board[i][g]='P'
                elif i==8:
                    if g==1 or g==8:
                        cls.board[i][g]='R'
                    elif g==2 or g==7:
                        cls.board[i][g]='N'
                    elif g==3 or g==6:
                        cls.board[i][g]='B'
                    elif g==4:
                        cls.board[i][g]='Q'
                    elif g==5:
                        cls.board[i][g]='K'
                        
    @classmethod
    def field_init_new(cls):
        for i in range(1,len(cls.board)):
            for g in range(1,len(cls.board[i])-1):
                if i==1:
                    if g==1 or g==8:
                        cls.board[i][g]='o'
                    elif g==2 or g==7:
                        cls.board[i][g]='a'
                    elif g==3 or g==6:
                        cls.board[i][g]='b'
                    elif g==4:
                        cls.board[i][g]='s'
                    elif g==5:
                        cls.board[i][g]='k'
                elif i==2:
                    cls.board[i][g]='p'
                elif i==7:
                    cls.board[i][g]='P'
                elif i==8:
                    if g==1 or g==8:
                        cls.board[i][g]='O'
                    elif g==2 or g==7:
                        cls.board[i][g]='A'
                    elif g==3 or g==6:
                        cls.board[i][g]='B'
                    elif g==4:
                        cls.board[i][g]='S'
                    elif g==5:
                        cls.board[i][g]='K'
            
        cls.white_kill=[]
        cls.black_kill=[]
        
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.b=type(self).board[9-self.y][self.x]
    
    def draw(self):
        type(self).board[9-self.y][self.x]=self.b
    
    def move_f(self,x2,y2):
        if type(self).board[9-y2][x2]!='.':
            if type(self).board[9-y2][x2].isupper():
                type(self).white_kill.append(type(self).board[9-y2][x2])
                type(self).board[9-self.y][self.x]='.'
                self.x=x2
                self.y=y2
                self.draw()
                
            elif type(self).board[9-y2][x2].islower():
                type(self).black_kill.append(type(self).board[9-y2][x2])
                type(self).board[9-self.y][self.x]='.'
                self.x=x2
                self.y=y2
                self.draw()
            else:
                print('Не туда воюешь!')
        else:
            type(self).board[9-self.y][self.x]='.'
            self.x=x2
            self.y=y2
            self.draw()
```
